# Build a recommender system by using cosine similarities score.
Dataset- book.csv

```python
import pandas as pd
import numpy as np
```

```python
#Read the data set
book_df = pd.read_csv('book.csv', encoding='latin-1')
```

```python
book_df
```

```python
book_df.info()
```

```python
book_data= book_df.iloc[:,1:]
```

```python
#Rename columns
book_data= book_data.rename({'User.ID': 'User_id','Book.Title':'Book_Title','Book.Rating':'Rating'},axis=1)
```

```python
# Sorting records by User_id
book_sort = book_data.sort_values('User_id')
book_sort
```

```python
#number of unique users in the dataset
len(book_sort.User_id.unique())
```

```python
#number of unique book titles in the dataset
len(book_sort.Book_Title.unique())
```

```python
# Converting long table format into wide format using Pivot Table

user_book = book_sort.pivot_table(index='User_id',columns='Book_Title',values='Rating').reset_index(drop=True)
user_book
```

```python
user_book.index = book_sort.User_id.unique()
```

```python
user_book
```

```python
#Impute NaNs with 0 values
user_book.fillna(0, inplace=True)
```

```python
user_book.head()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1- pairwise_distances( user_book.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.iloc[0:10,0:10]
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book_data.User_id.unique()
user_sim_df.columns = book_data.User_id.unique()
```

```python
user_sim_df.iloc[0:10,0:10]
```

```python
#fill diagonal values with zero
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)
```
